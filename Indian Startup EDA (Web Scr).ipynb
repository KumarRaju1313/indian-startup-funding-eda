#### 1. Webscraping the Data from trak.in Urs

```python
# Applying Link and checking a response and the page title
url = 'https://trak.in/india-startup-funding-investment-2015/'
response = requests.get(url)
print(response)
soup = BeautifulSoup(response.text)
```

```python
# creating the Headers for the table
tableHead = soup.thead
row_headers = []
for x in tableHead.find_all('tr'):
    for y in x.find_all('th'):
        row_headers.append(y.text)

row_headers1 = row_headers + ['index']
```

```python
# creating the body of the table
table_bodies = soup.find_all('tbody') # check all tbody elements
data_list = []  # Initialize an empty list to store the data rows

for tbody in table_bodies:  # Extract data rows from each <tbody>
    data_rows = tbody.find_all('tr')
    for row in data_rows:
        # Process each data row as needed
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        data_list.append(row_data)  # Append the row data as a list
```

```python
df = pd.DataFrame(data_list, columns=row_headers1)
df.to_csv("data_part1.csv")
df.head(10)
```

```python
# finding all url's in web page
url = 'https://trak.in/india-startup-funding-investment-2015/'
soup = BeautifulSoup(response.text)
funding_links = soup.find_all("h3")

for link in funding_links: # links in a single single formet
    url1 = link.find('a', {'data-wpel-link':'internal'})['href']
web_links = [] # taking an empty list to save links
for link in funding_links:
    url1 = link.find('a', {'data-wpel-link':'internal'})['href']
    web_links.append(url1)
# Unwanted Links
unwanted_links = ["https://trak.in/india-startup-funding-investment-2015/january-2015/", 
                  "https://trak.in/india-startup-funding-investment-2015/february-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/march-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/april-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/may-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/june-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/july-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/august-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/september-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/october-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/november-2015/",
                  "https://trak.in/india-startup-funding-investment-2015/december-2015/"]
# Filter out unwanted links
filtered_links = [link for link in web_links if link not in unwanted_links]
all_urls = filtered_links
```

```python
# Defining a function for scraping all URLs
def scrape_single_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    table_bodies = soup.find_all('tbody')
    for tbody in table_bodies:  # Extract data rows from each <tbody>
        data_rows = tbody.find_all('tr')
    return data_rows  # Return only data rows, excluding headers
```

```python
data_list = []  # Initialize an empty list to store the data rows

for url in all_urls:
    data_rows = scrape_single_url(url)  # Extract data rows from the URL
    # Exclude the first row if it's considered as header
    if data_rows and len(data_rows):  
        data_rows = data_rows[0:]
    for row in data_rows:
        # Process each data row as needed
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        data_list.append(row_data)  # Append the row data as a list

# Print the collected data
new_data = []  # Initialize an empty list to store the row data
for row in data_list:
    new_data.append(row)# Print the row data
new_data
```

```python
df1 = pd.DataFrame(new_data)
df1.to_csv("data_part2.csv")
df1.info()
```

```python
import csv

with open('data_part2.csv', 'r', newline='') as csvfile: # Read the CSV file
    reader = csv.DictReader(csvfile)
    rows = [row for row in reader]

sorted_rows = sorted(rows, key=lambda x: x['1'])  # Sort the rows based on the 'name' column

with open('data_part2.csv', 'w', newline='') as csvfile: # Write the sorted rows to a new CSV file
    fieldnames = reader.fieldnames
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in sorted_rows:
        writer.writerow(row)

print("CSV file sorted and saved successfully!")
pd.read_csv("data_part2.csv")
```
